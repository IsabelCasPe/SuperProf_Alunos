Resolução de equações diferenciais simples, como dP/dt= kP
 (crescimento exponencial) ou dP/dt= -kP (decaimento).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
# Equação diferencial: dP/dt = kP (crescimento exponencial)
def growth(P, t, k):
    return k * P
```

```python
# Parâmetros
k = 0.5  # Taxa de crescimento
P0 = 100  # População inicial
t = np.linspace(0, 10, 2000)  # Tempo
```

```python
# Resolver
P = odeint(growth, P0, t, args=(k,))
```

```python
# Plotar
plt.figure(figsize=(12, 8))
plt.plot(t, P, label=f"Crescimento (k = {k})")
plt.title("Crescimento Exponencial", fontsize=16)
plt.xlabel("Tempo", fontsize=12)
plt.ylabel("População", fontsize=12)
plt.grid(True)
plt.legend()
plt.show()
```
