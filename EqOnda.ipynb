# A equação da onda em uma corda, onde você verá a perturbação inicial se propagando ao longo da corda.
 A equação da onda em uma dimensão espacial
x e no tempo t é dada por:

∂^2u/∂t^2=c^2∂^2u/∂x^2

onde:
u(x,t) representa o deslocamento da onda em função do tempo e da posição,
c é a velocidade de propagação da onda.
Podemos resolver essa equação numericamente para ver a propagação de uma onda ao longo do tempo. Vou configurar o código em Python para simular uma onda em uma corda, usando uma solução numérica para visualizar a dinâmica da equação da onda.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

```python
# Parâmetros da equação da onda
L = 1.0          # comprimento da corda
c = 1.0          # velocidade de propagação da onda
dx = 0.01        # passo no espaço
dt = 0.005       # passo no tempo
nx = int(L / dx) # número de pontos no espaço
nt = 500         # número de passos no tempo
```

```python
# Malha espacial e inicialização das condições
x = np.linspace(0, L, nx)
u = np.zeros(nx)            # posição atual
u_prev = np.zeros(nx)       # posição anterior
u_next = np.zeros(nx)       # posição futura
```

```python
# Condição inicial: uma "perturbação" na posição inicial
u[int(nx/4):int(nx/2)] = 1
```

```python
# Parâmetro para o método numérico
C = (c * dt / dx)**2
```

```python
# Configuração do gráfico
fig, ax = plt.subplots()
line, = ax.plot(x, u, color='teal')
ax.set_ylim(-8.5, 8.5)
ax.set_title("Simulação da Equação da Onda em uma Corda")
ax.set_xlabel("Posição (x)")
ax.set_ylabel("Deslocamento (u)")

# Função de atualização para a animação
def update(frame):
    global u, u_prev, u_next
    # Atualização da solução para cada ponto no espaço (exceto as bordas)
    for i in range(1, nx - 1):
        u_next[i] = 2 * u[i] - u_prev[i] + C * (u[i + 1] - 2 * u[i] + u[i - 1])
    # Atualiza as posições para o próximo quadro
    u_prev, u, u_next = u, u_next, u_prev
    line.set_ydata(u)
    return line,

# Animação da propagação da onda
ani = FuncAnimation(fig, update, frames=nt, blit=True, interval=20, repeat=False)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Parâmetros da equação da onda
L = 1.0          # comprimento da corda
c = 1.0          # velocidade de propagação da onda
dx = 0.01        # passo no espaço
dt = 0.005       # passo no tempo
nx = int(L / dx) # número de pontos no espaço
nt = 500         # número de passos no tempo

# Malha espacial e inicialização das condições
x = np.linspace(0, L, nx)
u = np.zeros(nx)            # posição atual
u_prev = np.zeros(nx)       # posição anterior
u_next = np.zeros(nx)       # posição futura

# Condição inicial: uma "perturbação" na posição inicial
u[int(nx/4):int(nx/2)] = 1

# Parâmetro para o método numérico
C = (c * dt / dx)**2

# Configuração do gráfico
fig, ax = plt.subplots()
line, = ax.plot(x, u, color='teal')
ax.set_ylim(-5.5, 5.5)
ax.set_title("Simulação da Equação da Onda em uma Corda")
ax.set_xlabel("Posição (x)")
ax.set_ylabel("Deslocamento (u)")

# Função de atualização para a animação
def update(frame):
    global u, u_prev, u_next
    # Atualização da solução para cada ponto no espaço (exceto as bordas)
    for i in range(1, nx - 1):
        u_next[i] = 2 * u[i] - u_prev[i] + C * (u[i + 1] - 2 * u[i] + u[i - 1])
    # Atualiza as posições para o próximo quadro
    u_prev, u, u_next = u, u_next, u_prev
    line.set_ydata(u)
    return line,

# Animação da propagação da onda
ani = FuncAnimation(fig, update, frames=nt, blit=True, interval=20, repeat=False)
plt.show()
```
