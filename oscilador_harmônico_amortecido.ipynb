Esse código visualiza a trajetória do oscilador harmônico amortecido no plano de fase, mostrando como o deslocamento e a velocidade mudam ao longo do tempo com o efeito do amortecimento.
# Sistema do Oscilador Harmônico Amortecido
As equações para o oscilador harmônico amortecido são:

dx/dt=v
dv/dt=-w^2x-γv
onde:
x é o deslocamento,
v é a velocidade,
ω é a frequência natural do sistema (relacionada à força de restauração),
γ é o coeficiente de amortecimento (que reduz a amplitude ao longo do tempo).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
# Parâmetros do sistema do oscilador harmônico amortecido
omega = 1.0   # frequência natural
gamma = 0.1   # coeficiente de amortecimento

# Função que define o sistema do oscilador harmônico amortecido
def damped_oscillator(t, state):
    x, v = state
    dx_dt = v
    dv_dt = -omega**2 * x - gamma * v
    return [dx_dt, dv_dt]

# Condições iniciais para o deslocamento e velocidade
initial_state_osc = [1.0, 0.0]

# Intervalo de tempo e pontos de avaliação
t_span_osc = (0, 20)
t_eval_osc = np.linspace(t_span_osc[0], t_span_osc[1], 1000)

# Solução numérica do sistema
solution_osc = solve_ivp(damped_oscillator, t_span_osc, initial_state_osc, t_eval=t_eval_osc)
x_osc, v_osc = solution_osc.y

```

```python
# Visualização da trajetória no plano de fase
plt.figure(figsize=(10, 6))
plt.plot(x_osc, v_osc, color='teal', lw=1.5)
plt.title("Dinâmica do Oscilador Harmônico Amortecido no Plano de Fase")
plt.xlabel("Deslocamento (x)")
plt.ylabel("Velocidade (v)")
plt.grid(True)
plt.show()
```
