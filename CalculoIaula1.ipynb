```python
# Reimportar bibliotecas após o reset
import numpy as np
import matplotlib.pyplot as plt

# Dados para os gráficos
x = np.linspace(-10, 10, 400)  # Intervalo de x

# Função Linear: y = mx + b (exemplo com m=2, b=1)
y_linear = 2 * x + 1

# Função Quadrática: y = ax² + bx + c (exemplo com a=1, b=-2, c=1)
y_quadratic = 1 * x**2 - 2 * x + 1

# Função Constante: y = c (exemplo com c=3)
y_constant = np.full_like(x, 3)

# Gráfico 1: Função Linear
plt.figure(figsize=(8, 5))
plt.plot(x, y_linear, label="y = 2x + 1", color="blue")
plt.title("Função Linear: y = mx + b")
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()

# Gráfico 2: Função Quadrática
plt.figure(figsize=(8, 5))
plt.plot(x, y_quadratic, label="y = x² - 2x + 1", color="green")
plt.title("Função Quadrática: y = ax² + bx + c")
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()

# Gráfico 3: Função Constante
plt.figure(figsize=(8, 5))
plt.plot(x, y_constant, label="y = 3", color="red")
plt.title("Função Constante: y = c")
plt.xlabel("x")
plt.ylabel("y")
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.show()
```

```python
from pptx import Presentation
from pptx.util import Pt, Inches

# Criar a apresentação PowerPoint
presentation = Presentation()

# Slide 1: Enunciado do Exercício
slide = presentation.slides.add_slide(presentation.slide_layouts[1])
title = slide.shapes.title
content = slide.placeholders[1]
title.text = "Derivada com Passos Interativos"
content.text = (
    "Exercício:\n"
    "Derive a seguinte função:\n"
    "y = ln(x + √(x² + 4))"
)

# Slide 2: Passos Interativos
slide = presentation.slides.add_slide(presentation.slide_layouts[1])
title = slide.shapes.title
content = slide.placeholders[1]
title.text = "Passo a Passo da Derivação"

# Adicionando texto para os passos com explicações
steps = [
    "Passo 1: Identificar a função u(x):\n"
    "u(x) = x + √(x² + 4)",

    "Passo 2: Derivar u(x):\n"
    "u'(x) = 1 + (x / √(x² + 4))",

    "Passo 3: Usar a fórmula da derivada do logaritmo:\n"
    "y' = (1 / u(x)) * u'(x)",

    "Passo 4: Substituir u(x) e u'(x):\n"
    "y' = (1 / (x + √(x² + 4))) * (1 + (x / √(x² + 4)))"
]

top_position = 1.5  # Posição inicial do texto no slide
for step in steps:
    textbox = slide.shapes.add_textbox(Inches(0.5), Inches(top_position), Inches(9), Inches(0.5))
    text_frame = textbox.text_frame
    text_frame.text = step
    for paragraph in text_frame.paragraphs:
        paragraph.font.size = Pt(18)
    top_position += 0.8  # Ajusta a posição vertical para o próximo passo

# Salvar a apresentação
file_path = "/mnt/data/Passos_Interativos_Derivada.pptx"
presentation.save(file_path)

file_path
```
