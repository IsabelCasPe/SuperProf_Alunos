# sistema de Lotka-Volterra, mostrando as intera√ß√µes entre a popula√ß√£o de presas e predadores.
Este sistema modela a intera√ß√£o entre duas esp√©cies (presa e predador) e √© dado por:

dx/dt=ùõºùë•‚àíùõΩùë•ùë¶
dy/dt=ùõøùë•ùë¶-Œ≥y
onde:
x representa a popula√ß√£o da presa,

y representa a popula√ß√£o do predador,

Œ±, ùõΩ, ùõø e Œ≥ s√£o par√¢metros positivos.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
# Par√¢metros do sistema de Lotka-Volterra
alpha = 1.1  # taxa de crescimento da presa
beta = 0.4   # taxa de preda√ß√£o
delta = 0.1  # taxa de reprodu√ß√£o do predador em fun√ß√£o da presa
gamma = 0.4  # taxa de morte do predador
```

```python
# Fun√ß√£o do sistema de Lotka-Volterra
def lotka_volterra(t, state):
    x, y = state
    dx_dt = alpha * x - beta * x * y
    dy_dt = delta * x * y - gamma * y
    return [dx_dt, dy_dt]
```

```python
# Condi√ß√µes iniciais e integra√ß√£o
initial_state_lv = [10, 5]
t_span_lv = (0, 30)
t_eval_lv = np.linspace(t_span_lv[0], t_span_lv[1], 1000)
solution_lv = solve_ivp(lotka_volterra, t_span_lv, initial_state_lv, t_eval=t_eval_lv)
x_lv, y_lv = solution_lv.y
```

```python
# Visualiza√ß√£o das trajet√≥rias
plt.figure(figsize=(10, 6))
plt.plot(x_lv, y_lv, color='purple', lw=1.5)
plt.title("Din√¢mica do Sistema de Lotka-Volterra (Presa-Predador)")
plt.xlabel("Popula√ß√£o de Presa (x)")
plt.ylabel("Popula√ß√£o de Predador (y)")
plt.grid(True)
plt.show()
```

```python
# Configura√ß√£o do gr√°fico para o sistema de Lotka-Volterra
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(x_lv, y_lv, color='purple', lw=1.5)
ax.set_title("Din√¢mica do Sistema de Lotka-Volterra (Presa-Predador)")
ax.set_xlabel("Popula√ß√£o de Presa (x)")
ax.set_ylabel("Popula√ß√£o de Predador (y)")
ax.grid(True)

plt.show()
```
