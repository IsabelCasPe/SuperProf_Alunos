 Movimento Harmônico Amortecido (Equações Diferenciais de Segunda Ordem)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
# Equação: m * x'' + b * x' + k * x = 0
def harmonic(state, t, m, b, k):
    x, v = state
    dx_dt = v
    dv_dt = -(b/m) * v - (k/m) * x
    return [dx_dt, dv_dt]
```

```python
# Parâmetros
m = 1.0  # Massa
b = 0.5  # Coeficiente de amortecimento
k = 2.0  # Constante da mola
t = np.linspace(0, 20, 1000)
initial_state = [1.0, 0.0]  # Posição inicial 1, velocidade inicial 0
```

```python
# Resolver
solution = odeint(harmonic, initial_state, t, args=(m, b, k))
x = solution[:, 0]
```

```python
# Plotar
plt.figure(figsize=(12, 8))
plt.plot(t, x, label="Posição")
plt.title("Oscilador Harmônico Amortecido", fontsize=16)
plt.xlabel("Tempo (s)", fontsize=12)
plt.ylabel("Posição (m)", fontsize=12)
plt.grid(True)
plt.legend()
plt.show()
```
