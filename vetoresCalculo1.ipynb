```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def plot_vectors(vectors, colors, labels, title="Visualização de Vetores em R³"):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.set_xlim([-5, 5])
    ax.set_ylim([-5, 5])
    ax.set_zlim([-5, 5])
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title(title)

    for i, v in enumerate(vectors):
        ax.quiver(0, 0, 0, v[0], v[1], v[2], color=colors[i], label=labels[i], arrow_length_ratio=0.1)

    ax.legend()
    plt.show()

# Definição de vetores
u = np.array([1, 2, 3])
v = np.array([4, -1, 2])

# Soma de vetores
soma = u + v

# Produto escalar
produto_escalar = np.dot(u, v)

# Produto vetorial
produto_vetorial = np.cross(u, v)

# Exibir valores
print("u + v =", soma)
print("u . v =", produto_escalar)
print("u x v =", produto_vetorial)

# Plotar os vetores
plot_vectors([u, v, soma, produto_vetorial], ['r', 'b', 'g', 'purple'], ['u', 'v', 'u+v', 'u x v'])
```

```python
# Exercício 1: Determinar o domínio das funções

# Função 1: f(x,y) = ln(4 - x^2 - y^2)
# O domínio é definido para 4 - x^2 - y^2 > 0 -> x^2 + y^2 < 4 (interior de um círculo)

theta = np.linspace(0, 2*np.pi, 300)
x1 = 2 * np.cos(theta)
y1 = 2 * np.sin(theta)

# Função 2: g(x,y) = x / (y^2 - 1)
# O denominador não pode ser zero, então y ≠ ±1

x2 = np.linspace(-3, 3, 100)
y2_pos = np.ones(100)
y2_neg = -np.ones(100)

# Criando os gráficos dos domínios
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Domínio da função f(x,y)
axs[0].plot(x1, y1, color="b")
axs[0].fill_between(x1, y1, color="blue", alpha=0.2)
axs[0].set_xlim([-3, 3])
axs[0].set_ylim([-3, 3])
axs[0].set_title(r"Domínio de $f(x,y) = \ln(4 - x^2 - y^2)$")
axs[0].set_xlabel("x")
axs[0].set_ylabel("y")

# Domínio da função g(x,y)
axs[1].plot(x2, y2_pos, "--r", label="y = 1")
axs[1].plot(x2, y2_neg, "--r", label="y = -1")
axs[1].set_xlim([-3, 3])
axs[1].set_ylim([-3, 3])
axs[1].set_title(r"Domínio de $g(x,y) = \frac{x}{y^2 - 1}$")
axs[1].set_xlabel("x")
axs[1].set_ylabel("y")
axs[1].legend()

plt.show()
```
