```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
# Crear un grafo para representar el rbol
G = nx.DiGraph()

# Nodos iniciales (caras del dado)
caras_dado = [1, 2, 3, 4, 5, 6]
for cara in caras_dado:
    G.add_edge("Dado", f"{cara}")

# Determinar los eventos
def resultado(cara):
    if cara in [1, 3, 5] and cara > 4:
        return "Mara y Juan"
    elif cara in [1, 3, 5]:
        return "Mara"
    elif cara > 4:
        return "Juan"
    else:
        return "Nadie"

# Agregar las ramas de resultados
for cara in caras_dado:
    evento = resultado(cara)
    G.add_edge(f"{cara}", f"{evento}")

# Dibujar el grafo (rbol)
pos = {
    "Dado": (0, 2),
    "1": (-2, 1), "2": (-1, 1), "3": (0, 1), "4": (1, 1), "5": (2, 1), "6": (3, 1),
    "Mara": (-2, 0), "Nadie": (-1, 0), "Mara y Juan": (2, 0), "Juan": (3, 0)
}

plt.figure(figsize=(7, 6))
nx.draw(G, pos, with_labels=True, node_size=2000, node_color="skyblue", font_size=10, font_weight="bold")
plt.title("Diagrama de rbol: Probabilidades de Mara y Juan")
plt.show()
```

Explicacin del Diagrama del rbol
el punto inicial es el lanzamiento del dado.
las 6 caras del dado se ramificam como resultados individuais: 1, 2, 3, 4, 5, 6.
A partir de cada resultado:
1, 3, 5  Vitria de Mara (nmeros mpares).
5, 6  Vitria de Juan (nmeros > 4).
El nmero 5 pertence a Mara y Juan, pues es mpar y mayor que 4.
2 e 4  No son favorables a nadie.
