e^ix ; percorre o círculo unitário à medida que x varia de 0 a
2π. A parte real é o eixo x, a parte imaginária o eixo y

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Valores de x de 0 a 2*pi
x = np.linspace(0, 2 * np.pi, 500)

# Parte real e imaginária da equação de Euler
real = np.cos(x)
imag = np.sin(x)
```

```python
plt.figure(figsize=(12,12))
plt.plot(real, imag, label=r'$e^{ix} = \cos(x) + i\sin(x)$')
plt.xlabel('Parte Real (cos(x))')
plt.ylabel('Parte Imaginária (sin(x))')
plt.title('Círculo Unitário — Representação de $e^{ix}$')
plt.grid(True)
plt.axhline(0, color='gray', linestyle='--', linewidth=0.5)
plt.axvline(0, color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.axis('equal')  # para manter o círculo redondo
plt.show()
```
