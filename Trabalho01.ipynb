Cálculo de trabalho usando integrais, W=∫Fdx , aplicado a um sistema dinâmico.

Calcular o trabalho necessário para mover um objeto em um campo gravitacional (ex.: lançar um foguete) e simular sua trajetória sob gravidade.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Força gravitacional: F = GMm / r^2
G = 6.67430e-11  # Constante gravitacional
M = 5.972e24     # Massa da Terra (kg)
m = 5000         # Massa do objeto (kg)
r0 = 6.371e6     # Raio da Terra (m)
rf = 1e7         # Altura final (m)
```

```python
# Trabalho: W = integral de F dr de r0 a rf
r = np.linspace(r0, rf, 1000)
F = G * M * m / r**2
W = np.trapezoid(F, r)  # Integração numérica using np.trapezoid
```

```python
# Trajetória simplificada (assumindo velocidade inicial)
v0 = 1000  # Velocidade inicial (m/s)
t = np.linspace(0, 30, 1000)
g = 9.81  # Aceleração da gravidade (m/s^2)
h = v0 * t - 0.5 * g * t**2  # Altura (desprezando variação de g por simplicidade)
```

```python
# Plotar
plt.figure(figsize=(12, 8))
plt.plot(t, h, label="Trajetória")
plt.title(f"Trajetória e Trabalho (W = {W:.2e} J)", fontsize=12)
plt.xlabel("Tempo (s)", fontsize=12)
plt.ylabel("Altura (m)", fontsize=12)
plt.grid(True)
plt.legend()
plt.show()
```
