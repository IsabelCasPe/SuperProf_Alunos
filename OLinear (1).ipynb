```python
import numpy as np
import matplotlib.pyplot as plt

# Criando o espaço das variáveis
x1 = np.linspace(0, 7, 200)

# Restrições
x2_1 = (60 - 10*x1) / 12  # 10x1 + 12x2 <= 60
x2_2 = (6 - 2*x1)         # 2x1 + x2 <= 6

# Região factível
plt.figure(figsize=(8,6))
plt.plot(x1, x2_1, label=r'$10x_1 + 12x_2 \leq 60$', color='blue')
plt.plot(x1, x2_2, label=r'$2x_1 + x_2 \leq 6$', color='red')

# Limitando a região viável
plt.fill_between(x1, np.minimum(x2_1, x2_2), where=(x1 >= 0), color='gray', alpha=0.3)

# Pontos extremos
points = np.array([[0,5], [3,0], [6,0], [6/7, 30/7]])

# Plotando os pontos viáveis
for point in points:
    plt.scatter(point[0], point[1], color='black')
    plt.text(point[0], point[1], f'({point[0]:.2f}, {point[1]:.2f})', fontsize=12, verticalalignment='bottom')

# Destacando o ponto ótimo
plt.scatter(6, 0, color='green', s=100, label='Máximo em (6,0)')

# Configurações do gráfico
plt.xlim(0, 7)
plt.ylim(0, 7)
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.legend()
plt.title('Região Viável e Solução Ótima')

# Exibir gráfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir as restrições
x1 = np.linspace(0, 10, 200)

# Restrições
x2_1 = (80 - 20*x1) / 10  # 20x1 + 10x2 ≤ 80
x2_2 = 5 - x1  # x1 + x2 ≥ 5

# Criar a área viável
plt.figure(figsize=(8,6))
plt.plot(x1, x2_1, label=r'$20x_1 + 10x_2 \leq 80$', linestyle='dashed')
plt.plot(x1, x2_2, label=r'$x_1 + x_2 \geq 5$', linestyle='dashed')

# Preenchendo a área viável
x2_fill = np.maximum(0, np.minimum(x2_1, x2_2))
plt.fill_between(x1, 0, x2_fill, alpha=0.3, color='gray')

# Definir pontos de interseção manualmente
A = (0, 5)  # Interseção do eixo y com x1 + x2 = 5
B = (4, 1)  # Interseção das duas restrições
C = (4, 0)  # Interseção do eixo x com 20x1 + 10x2 ≤ 80

# Marcar pontos de interseção
plt.scatter(*A, color='red', label='A (0,5)')
plt.scatter(*B, color='blue', label='B (4,1)')
plt.scatter(*C, color='green', label='C (4,0)')

# Configurações do gráfico
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.xlabel(r'$x_1$')
plt.ylabel(r'$x_2$')
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.legend()
plt.title("Região viável e pontos de interseção")

# Exibir gráfico
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir as restrições
x1 = np.linspace(0, 12, 400)

# Restrição 1: 5x1 + 2x2 <= 40  -> x2 <= (40 - 5x1) / 2
x2_1 = (40 - 5*x1) / 2

# Restrições de limites
x2_limit = np.full_like(x1, 15)  # x2 <= 15
x1_limit = np.full_like(x1, 10)  # x1 <= 10

# Definir a região viável
plt.figure(figsize=(8,6))
plt.plot(x1, x2_1, label=r'$5x_1 + 2x_2 = 40$', color='blue')
plt.axhline(y=15, color='green', linestyle='--', label=r'$x_2 = 15$')
plt.axvline(x=10, color='purple', linestyle='--', label=r'$x_1 = 10$')

# Preenchendo a região viável
x_fill = np.linspace(0, 10, 400)
x2_fill = np.minimum((40 - 5*x_fill) / 2, 15)  # Região válida entre as restrições
plt.fill_between(x_fill, 0, x2_fill, color='gray', alpha=0.3)

# Plotar os pontos de vértices da região viável
vertices = np.array([[0, 0], [0, 15], [8, 0], [8, 5]])
plt.scatter(vertices[:, 0], vertices[:, 1], color='red', zorder=3)

# Destacando a solução ótima
plt.scatter(8, 5, color='gold', edgecolors='black', s=100, label="Solução Ótima (8,5)")

# Configurar eixos e legenda
plt.xlim(0, 12)
plt.ylim(0, 16)
plt.xlabel(r'$x_1$ (Mesas)')
plt.ylabel(r'$x_2$ (Cadeiras)')
plt.title('Solução Gráfica do Problema de Otimização - Marcenaria')
plt.legend()
plt.grid()

# Exibir o gráfico
plt.show()
```
