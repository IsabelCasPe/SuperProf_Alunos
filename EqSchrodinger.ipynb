Resolve a equa√ß√£o de Schr√∂dinger no po√ßo quadrado infinito para os primeiros tr√™s estados de energia.
Plota as fun√ß√µes de onda ‚à£ùúì‚à£^2

  para esses estados. O gr√°fico mostra a probabilidade de encontrar a part√≠cula em diferentes posi√ß√µes dentro do po√ßo.
A equa√ß√£o de Schr√∂dinger √© fundamental na mec√¢nica qu√¢ntica e descreve a evolu√ß√£o temporal do estado qu√¢ntico de uma part√≠cula. Vamos come√ßar com a equa√ß√£o de Schr√∂dinger independente do tempo em uma dimens√£o, que √© usada para descrever o comportamento da fun√ß√£o de onda
ùúì(ùë•) de uma part√≠cula com energia constante.

Equa√ß√£o de Schr√∂dinger Independente do Tempo
A forma da equa√ß√£o em uma dimens√£o √©:

‚àí‚Ñè^2/2ùëö*ùëë^2ùúì/ùëëùë•^2+ùëâ(ùë•)ùúì=ùê∏ùúì

onde:

ùúì(ùë•) √© a fun√ß√£o de onda da part√≠cula,
ùëâ(ùë•) √© o potencial que a part√≠cula experimenta,
ùê∏ √© a energia da part√≠cula,
‚Ñè √© a constante de Planck reduzida,
ùëö √© a massa da part√≠cula.
Para exemplificar, podemos usar um po√ßo quadrado infinito como potencial, onde
ùëâ(ùë•)=0 dentro de um intervalo 0<ùë•<ùêø e ùëâ(ùë•)=‚àû
V(x)=‚àû fora dele. Essa configura√ß√£o nos permite calcular e visualizar os estados estacion√°rios (os modos de vibra√ß√£o) da part√≠cula dentro do po√ßo.

Vou configurar o c√≥digo para resolver numericamente a equa√ß√£o de Schr√∂dinger e visualizar os modos estacion√°rios.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigh
```

```python
# Par√¢metros f√≠sicos
hbar = 1.0    # Constante de Planck reduzida
m = 1.0       # Massa da part√≠cula
L = 1.0       # Largura do po√ßo

# Discretiza√ß√£o espacial
nx = 1000                      # N√∫mero de pontos no espa√ßo
dx = L / (nx - 1)              # Intervalo de espa√ßamento
x = np.linspace(0, L, nx)      # Posi√ß√£o ao longo do eixo x

# Defini√ß√£o do potencial: Po√ßo quadrado infinito
V = np.zeros(nx)               # Potencial dentro do po√ßo (V(x) = 0)

# Constru√ß√£o da matriz Hamiltoniana
diag = np.full(nx, 2.0)        # Diagonal principal
off_diag = np.full(nx - 1, -1.0)  # Diagonal secund√°ria
H = (-hbar**2 / (2 * m * dx**2)) * (np.diag(diag) + np.diag(off_diag, -1) + np.diag(off_diag, 1)) + np.diag(V)

# C√°lculo dos autovalores e autovetores
energies, wavefunctions = eigh(H)  # Resolver H psi = E psi

# Seleciona os primeiros tr√™s estados
num_states = 3
fig, ax = plt.subplots(num_states, 1, figsize=(10, 8))
x_vals = x / L  # Normaliza√ß√£o da posi√ß√£o entre 0 e 1

for i in range(num_states):
    ax[i].plot(x_vals, wavefunctions[:, i]**2, label=f"Estado {i+1}, E = {energies[i]:.2f}")
    ax[i].set_ylabel(f"|œà|¬≤")
    ax[i].legend()
    ax[i].grid(True)

plt.xlabel("Posi√ß√£o normalizada (x/L)")
plt.suptitle("Fun√ß√µes de Onda dos Primeiros Estados no Po√ßo Quadrado Infinito")
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
