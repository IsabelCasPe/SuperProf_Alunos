Séries de Fourier e Ondas (Aproximação de Funções)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Função para calcular a série de Fourier de uma onda quadrada
def fourier_square(t, N):
    f = np.zeros_like(t)
    for n in range(1, N+1, 2):  # Apenas termos ímpares
        f += (4 / (np.pi * n)) * np.sin(n * 2 * np.pi * t)
    return f
```

```python
# Parâmetros
t = np.linspace(0, 2, 1000)
N_terms = 10  # Número de termos na série
```

```python
# Calcular
wave = fourier_square(t, N_terms)
```

```python
# Plotar
plt.figure(figsize=(12, 7))
plt.plot(t, wave, label=f"Série de Fourier ({N_terms} termos)")
plt.title("Aproximação de Onda Quadrada com Séries de Fourier", fontsize=16)
plt.xlabel("Tempo", fontsize=12)
plt.ylabel("Amplitude", fontsize=12)
plt.grid(True)
plt.legend()
plt.show()
```
