```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Criar um espaço para múltiplos gráficos
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Tempo para o eixo x
t = np.linspace(0, 10, 100)

# 1. MRU - Gráfico de posição x tempo
v_mru = 5  # Velocidade constante
x_mru = v_mru * t
axes[0, 0].plot(t, x_mru, label="x = v * t", color="blue")
axes[0, 0].set_title("Movimento Retilíneo Uniforme (MRU)")
axes[0, 0].set_xlabel("Tempo (s)")
axes[0, 0].set_ylabel("Posição (m)")
axes[0, 0].legend()
axes[0, 0].grid()

# 2. MRUV - Gráfico de posição x tempo
a_mruv = 2  # Aceleração constante
x_mruv = 2 * t + 0.5 * a_mruv * t**2
axes[0, 1].plot(t, x_mruv, label="x = x₀ + v₀t + ½at²", color="green")
axes[0, 1].set_title("Movimento Retilíneo Uniformemente Variado (MRUV)")
axes[0, 1].set_xlabel("Tempo (s)")
axes[0, 1].set_ylabel("Posição (m)")
axes[0, 1].legend()
axes[0, 1].grid()

# 3. MRUV - Gráfico de velocidade x tempo
v_mruv = 2 + a_mruv * t
axes[1, 0].plot(t, v_mruv, label="v = v₀ + at", color="red")
axes[1, 0].set_title("Velocidade vs Tempo (MRUV)")
axes[1, 0].set_xlabel("Tempo (s)")
axes[1, 0].set_ylabel("Velocidade (m/s)")
axes[1, 0].legend()
axes[1, 0].grid()

# 4. Interpretação da Derivada - x(t), v(t) e a(t)
x_t = t**2  # Exemplo genérico
v_t = np.gradient(x_t, t)  # Derivada numérica da posição
a_t = np.gradient(v_t, t)  # Derivada numérica da velocidade

axes[1, 1].plot(t, x_t, label="x(t)", color="blue")
axes[1, 1].plot(t, v_t, label="v(t) = dx/dt", color="red", linestyle="dashed")
axes[1, 1].plot(t, a_t, label="a(t) = dv/dt", color="green", linestyle="dotted")
axes[1, 1].set_title("Interpretação da Derivada na Cinemática")
axes[1, 1].set_xlabel("Tempo (s)")
axes[1, 1].legend()
axes[1, 1].grid()

# Ajustar layout e exibir
plt.tight_layout()
plt.show()
```
